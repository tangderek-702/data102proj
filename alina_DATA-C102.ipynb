```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("2022_US_Region_Mobility_Report.csv")
```

```python
king_df = df[(df["sub_region_1"] == "Washington") & (df["sub_region_2"] == "King County")].iloc[:, 8:15]
king_df
```

## Visualization 1

```python
king_df.plot(x="date", 
             y=[
                 "retail_and_recreation_percent_change_from_baseline", 
                 "grocery_and_pharmacy_percent_change_from_baseline", 
                 "parks_percent_change_from_baseline", 
                 "transit_stations_percent_change_from_baseline",
                 "workplaces_percent_change_from_baseline",
                 "residential_percent_change_from_baseline"
             ],
             figsize = (15, 8))
plt.title("Mobility Metric by Time (2022)")
plt.xlabel("Time (2022)")
plt.ylabel("Relative Change from Baseline");
```

```python
fuel_prices = pd.read_csv("Weekly_Seattle_WA_All_Grades_All_Formulations_Retail_Gasoline_Prices.csv")
fuel_prices = fuel_prices.loc[::-1].reset_index(drop=True)
fuel_prices
```

## Visualization 2

```python
fuel_prices.plot(x="Date", y="Fuel_Price")
plt.title("Fuel Prices in King County (WA) in 2022")
plt.xlabel("Date")
plt.ylabel("Fuel Price (USD)");
```

## Visualization 3

```python
days = []
for i in range(len(king_df)):
    if i % 7 == 0:
        days.append("Saturday")
    elif i % 7 == 1:
        days.append("Sunday")
    elif i % 7 == 2:
        days.append("Monday")
    elif i % 7 == 3:
        days.append("Tuesday")
    elif i % 7 == 4:
        days.append("Wednesday")
    elif i % 7 == 5:
        days.append("Thursday")
    elif i % 7 == 6:
        days.append("Friday")
king_df["Day"] = days
king_df[31:60].plot(x="date", 
             y=[
                 "retail_and_recreation_percent_change_from_baseline", 
                 "grocery_and_pharmacy_percent_change_from_baseline", 
                 "parks_percent_change_from_baseline", 
                 "transit_stations_percent_change_from_baseline",
                 "workplaces_percent_change_from_baseline",
                 "residential_percent_change_from_baseline"
             ],
             figsize = (15, 8))
```
